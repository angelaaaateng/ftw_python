## Train Test Split

Let's split up the data into a training set and a test set!

````python
from sklearn.model_selection import train_test_split

# we discussed the importance of splitting our dataset, now we will apply SKlearn's data split method
````

To learn more about the `sklearn.model_selection` library, check the documentation <a href='https://scikit-learn.org/stable/modules/cross_validation.html'>here</a>

````python
X = df.drop('Kyphosis',axis=1)
# removes the column Kyphosis from the DataFrame df to save it as the target variable
y = df['Kyphosis']
````

````python
X
````

````python
y
````

````python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state =42)
# 42 sets a fixed random seed to ensure reproducibility.Every time you run this code, the same data is split into training and testing sets.
````

````python
X_train.head()
````

````python
X_test.head()
````

````python
y_train.head()
````

````python
y_test.head()
````

## Decision Trees

We'll start just by training a single decision tree.

````python
from sklearn.tree import DecisionTreeClassifier
````

````python
dtree = DecisionTreeClassifier()
````

````python
dtree
````

````python
dtree.fit(X_train,y_train)
````

## Prediction and Evaluation

Let's evaluate our decision tree.

````python
predictions = dtree.predict(X_test)
# we're predicting off our test set
````

````python
predictions
````

````python
X_test.shape
````

````python
len(predictions)
````

````python
from sklearn.metrics import classification_report, confusion_matrix
# we'll talk about this more next week, this is just for demonstration purposes
````

````python
print(classification_report(y_test,predictions))
# score / evaluate based on our test data and then on our predictions
# the goal is to minimize the difference between the actual result and then our predictions
````

````python
print(confusion_matrix(y_test,predictions))

# we're mislabeling about 7 in our test set
# (tn, fp, fn, tp)

# [[(0,0), (0,1)
#   (1,0), (1,1)]]
````

````python
from sklearn.metrics import accuracy_score
````

````python
accuracy_score(y_test, predictions)
````

## Random Forests

Now let's compare the decision tree model to a random forest.

````python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(n_estimators=100, random_state=42)
rfc
````

````python
rfc.fit(X_train, y_train)
````

````python
rfc_pred = rfc.predict(X_test)
rfc_pred
````

````python
print(confusion_matrix(y_test,rfc_pred))
# how does our performance compare to a decision tree?
````

````python
print(classification_report(y_test,rfc_pred))
````

````python
accuracy_score(y_test, rfc_pred)
````

````python
features = list(df.columns[1:])
features
````

````python
tree = rfc.estimators_[0]
dot_data = StringIO()
export_graphviz(tree, out_file=dot_data, feature_names=features,filled=True,rounded=True)
graph = pydot.graph_from_dot_data(dot_data.getvalue())

graph
````
