## Describing our dataset with numbers

### What is the overall sex distribution of primary educators?

Based on the metadata table, we can check the code `SE.PRM.TCHR.FE.ZS` 

Here we'll likely look at average % of females, ranges among the countries, among the years

```python
world_df['Series Code'].value_counts()
```

```python
primary_educators_female = world_df[world_df['Series Code'] == 'SE.PRM.TCHR.FE.ZS']
primary_educators_female
```

```python
primary_educators_female.info()
```

```python
# Assumption: we can replace 'null' values with 0, for simplification 

primary_educators_female  = primary_educators_female.replace("..", "0")
primary_educators_female
```

```python
# primary_educators_female.columns
cols_to_avg = [
       '2010 [YR2010]', '2011 [YR2011]', '2012 [YR2012]', '2013 [YR2013]',
       '2014 [YR2014]', '2015 [YR2015]', '2016 [YR2016]', '2017 [YR2017]',
       '2018 [YR2018]', '2019 [YR2019]', '2020 [YR2020]']
```

```python
primary_educators_female[cols_to_avg] = primary_educators_female[cols_to_avg].apply(pd.to_numeric, errors='coerce')
primary_educators_female[cols_to_avg].dtypes
```

```python
primary_educators_female.describe()
```

```python
# Average % of females by country (row average)
primary_educators_female = primary_educators_female[['Country Name',
       '2010 [YR2010]', '2011 [YR2011]', '2012 [YR2012]', '2013 [YR2013]',
       '2014 [YR2014]', '2015 [YR2015]', '2016 [YR2016]', '2017 [YR2017]',
       '2018 [YR2018]', '2019 [YR2019]', '2020 [YR2020]']]
primary_educators_female
```

```python
primary_educators_female.mean()
```

```python
primary_educators_female.set_index('Country Name', inplace=True)
```

```python
primary_educators_female.mean(axis=1)
```
